#### Get the subtotal 

```python
# Function: Search value in dataframe, and find the next n non-empty cell
def next_non_empty(df, value, dim, n):
    mask = df.applymap(lambda x: str(value) in str(x)).to_numpy()
    if mask.any():
        r, c = np.argwhere(mask)[0]
    else:
        return df.shape[0]+1, df.shape[1]+1
    if dim == "row":
        for row in range(r+n, df.shape[0]):
            if df.iloc[row, c] is not np.nan: 
                return row, c
                break
    if dim == "col": 
        for col in range(c+n, df.shape[1]):
            if df.iloc[r, col] is not np.nan:
                return r, col
                break
```

```python
total = PO.iloc[next_non_empty(PO, "OrderTotal", "col", 1)]
try: CA_total = PO.iloc[next_non_empty(PO, "加拿大总数量", "row", 2)]
except (KeyError, IndexError): CA_total = 0
try: UK_total = PO.iloc[next_non_empty(PO, "英国总数量", "row", 2)]
except (KeyError, IndexError): UK_total = 0
try: DE_total = PO.iloc[next_non_empty(PO, "德国总数量", "row", 2)]
except (KeyError, IndexError): DE_total = 0
try: CN_total = PO.iloc[next_non_empty(PO, "中国总数量", "row", 2)]
except (KeyError, IndexError): CN_total = 0
destination = PO[PO.applymap(lambda x: '总数量' in str(x))].stack().tolist()
print(total, CA_total, UK_total, DE_total, CN_total, destination)
```
