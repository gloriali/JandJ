Search value in dataframe, and find the next non-empty cell.    

```python
def next_non_empty(df, value, dim):
    mask = df.applymap(lambda x: str(value) in str(x)).to_numpy()
    if mask.any():
        r, c = np.argwhere(mask)[0]
    else:
        return df.shape[0]+1, df.shape[1]+1
    if dim == "row":
        for row in range(r+1, df.shape[0]):
            if df.iloc[row, c] is not np.nan: 
                return row, c
                break
    if dim == "col": 
        for col in range(c+1, df.shape[1]):
            if df.iloc[r, col] is not np.nan:
                return r, col
                break
```

```python
cat = PO.iloc[next_non_empty(PO, "PO#", "col")].split("-")[2]
name = PO.iloc[next_non_empty(PO, "Category", "col")]
total = PO.iloc[next_non_empty(PO, "OrderTotal", "col")]
CA_total = PO.iloc[next_non_empty(PO, "CA Total Qty", "row")]
UK_total = PO.iloc[next_non_empty(PO, "UK Total Qty", "row")]
DE_total = PO.iloc[next_non_empty(PO, "DE Total Qty", "row")]
try: CN_total = PO.iloc[next_non_empty(PO, "CN Total Qty", "row")]
except (KeyError, IndexError): CN_total = ""
print(cat, name, total, CA_total, UK_total, DE_total, CN_total)
```
